```python
import random 
import matplotlib.pyplot as plt
```

```python
random_number=[random.uniform(0,1) for _ in range(10000)]
#print(random_number)
plt.hist(random_number,bins=50)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('unform sampling')
plt.show()
def Uniform_to_Bernoulli(random_numbers,probality):
    bernoulli_sampels=[]
    for i in random_numbers:
        if i > probality:
            bernoulli_sampels.append(1)
        else:
            bernoulli_sampels.append(0)
    return bernoulli_sampels            
bernoulli_numbers=Uniform_to_Bernoulli(random_number,.3)
#print (bernoulli_numbers)
success_rate=sum(bernoulli_numbers)
failure_rate=len(bernoulli_numbers)-success_rate
print(success_rate)
print(failure_rate)
plt.bar(['success','failure'],[success_rate,failure_rate])
plt.xlabel('Outcome')
plt.ylabel('frequency')
plt.title('Uniform to Bernoulli sampling')
plt.show()
```

```python
def genarte_sampels(a,b):
    samples=[]
    for _ in range(10000):
        random_number=random.random()
        sample=a+(b-a)*random_number
        samples.append(sample)
    return samples

a=1 
b=5
genarte_sampels_data=genarte_sampels(a,b)
#print('genarate samples',genarte_sampels_data)
plt.hist(genarte_sampels_data,bins=20)
plt.xlabel('Samples')
plt.ylabel('frequency')
plt.title('sampling data')
plt.show()
```

```python
class MersenneTwister:
    def __init__(self, seed):
        self.index = 0
        self.MT = [0] * 19937
        self.MT[0] = seed

        for i in range(1, 19937):
            self.MT[i] = (1812433253 * (self.MT[i - 1] ^ (self.MT[i - 1] >> 30)) + i) & 0xFFFFFFFF

    def extract_number(self):
        if self.index == 0:
            self.generate_numbers()

        y = self.MT[self.index]
        y ^= (y >> 11)
        y ^= ((y << 7) & 0x9D2C5680)
        y ^= ((y << 15) & 0xEFC60000)
        y ^= (y >> 18)

        self.index = (self.index + 1) % 19937
        return y / 4294967296.0

    def generate_numbers(self):
        for i in range(19937):
            y = (self.MT[i] & 0x80000000) + (self.MT[(i + 1) % 19937] & 0x7FFFFFFF)
            self.MT[i] = self.MT[(i + 397) % 19937] ^ (y >> 1)

            if y % 2 != 0:
                self.MT[i] ^= 0x9908B0DF

seed = 1234
mt = MersenneTwister(seed)

for _ in range(10):
    random_float = mt.extract_number()
    print(random_float)
```
